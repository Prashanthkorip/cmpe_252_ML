```python
import matplotlib.pyplot as plt
import numpy as np
from math import comb
```

```python
def func(x):
    return comb(11, 8) * (x**(8)) * ((1-x)**(11-8))
```

```python
x = np.arange(0, 1.1, 0.1)
l = [func(i) for i in x]
plt.plot(x, l, 'r')
plt.xticks(x)
plt.xlabel('theta')
plt.ylabel('Likehood')
# draw a line at the max of the graph to the bottom x axis
plt.axvline(x[np.argmax(l)], color='k', linestyle='--')
# get the value where the line is drawn on the x axis

plt.show()
```
